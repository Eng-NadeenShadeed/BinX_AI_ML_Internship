### Step 4: Data Cleaning Check

### Step 5: Categorizing Performance (Custom Function)

We write a reusable function - a Day 2 Python fundamentals skill - to 
turn the numeric GPA into a readable performance category.

```python
def categorize_performance(gpa):
    """Categorize a student's GPA into a performance band."""
    if gpa >= 3.5:
        return "Excellent"
    elif gpa >= 2.5:
        return "Good"
    elif gpa >= 1.5:
        return "Average"
    else:
        return "At Risk"

df["PerformanceCategory"] = df["GPA"].apply(categorize_performance)
print(df["PerformanceCategory"].value_counts())
```

### Step 6: Standardizing GPA with NumPy

Using NumPy, we compute the z-score of each student's GPA - how many 
standard deviations above/below the class average each student is. This 
is the required "NumPy-derived feature" for the mini-notebook.

```python
import numpy as np

gpa_mean = np.mean(df["GPA"])
gpa_std = np.std(df["GPA"])
df["GPA_zscore"] = (df["GPA"] - gpa_mean) / gpa_std

print("Class average GPA:", gpa_mean)
print("Standard deviation:", gpa_std)
print(df[["GPA", "GPA_zscore"]].head())
```

### Step 7: Does Tutoring or Parental Support Matter?

Grouping by Tutoring and Parental Support to see whether they're 
associated with higher GPA.

```python
print("Average GPA by Tutoring:")
print(df.groupby("Tutoring")["GPA"].mean())

print("\nAverage GPA by Parental Support level:")
print(df.groupby("ParentalSupport")["GPA"].mean())
```

### Step 8: Visualizing the Data

```python
# Histogram - distribution of GPA
plt.figure()
plt.hist(df["GPA"], bins=20, color="steelblue")
plt.title("Distribution of Student GPA")
plt.xlabel("GPA")
plt.ylabel("Number of Students")
plt.show()
```

**What this reveals:** Most students cluster in the 1.0-2.5 GPA range, 
with relatively few reaching "Excellent" (3.5+) - matching the category 
counts from Step 5.

```python
# Scatter - Study time vs GPA
plt.figure()
plt.scatter(df["StudyTimeWeekly"], df["GPA"], alpha=0.4, s=10)
plt.title("Study Time vs. GPA")
plt.xlabel("Study Time per Week (hours)")
plt.ylabel("GPA")
plt.show()
```

**What this reveals:** There's a visible upward trend - students who 
study more hours per week tend to have higher GPAs, though the 
relationship isn't perfectly linear (lots of spread/noise).

```python
# Bar chart - average GPA by tutoring status
tutoring_avg = df.groupby("Tutoring")["GPA"].mean()
plt.figure()
plt.bar(["No Tutoring", "Tutoring"], tutoring_avg.values, color=["gray", "steelblue"])
plt.title("Average GPA: Tutoring vs. No Tutoring")
plt.xlabel("Tutoring Status")
plt.ylabel("Average GPA")
plt.show()
```

**What this reveals:** Students with tutoring have a noticeably higher 
average GPA than those without - consistent with the groupby result in 
Step 7.
