NumPy Array Manipulation

Step 1: Create a 2D array of shape (4, 4) filled with values 1-16 and print its shape and dtype.

```python
import numpy as np 

 # Create a 4x4 array with values from 1 to 16
arr = np.arange(1,17).reshape(4,4)
# Print the original array
print(arr)
# print the shape of the array
print("\nShape: " , arr.shape)

# Print array data type
print("\nData type: " , arr.dtype)

```

Step 2: Use slicing to extract the second column and the last row.

```python
# Extract the second column and the last row
seconed_column = arr[:, 1]

last_row = arr[-1, :]

print("Second column:", seconed_column)
print("Last row:", last_row)
```

Step 3: Use a boolean mask to select all values greater than the array's mean.

```python
# Calculate the mean of the array
mean_value = arr.mean()
# Filter the array to get values greater than the mean
filtered_arr = arr[arr > mean_value]
print("Mean value:", mean_value)
print("Filtered array (values greater than mean):", filtered_arr)
```

Step 4: Add a 1D row array to every row of a 2D array using broadcasting, and verify the result manually.

```python

row = np.array([10, 20, 30, 40])
# Add the row array to evry row of the matrix
result = arr  + row
print (result)
```
