## Step 1
Load the CSV dataset into a DataFrame and report its shape, columns, and dtypes.

```python
import pandas as pd

df = pd.read_csv("data/melb_data.csv")

print("Shape: " , df.shape)
print("\nColumn: " , list(df.columns))
print("\nData Types:\n " , df.dtypes)
```

## Step 2
Count missing values per column and document a justification for handling them.

```python
print ("missing value per column: ")
print(df.isnull().sum())

print("\n duplicate rows: " , df.duplicated().sum())
```

## Step 3
Filter to a meaningful subset and describe the finding.

 **Finduing:**  
I filtered the dataset to include only houses (Type == 'h') with a **price** greater than $1,000,000. Then, I grouped the filtered results by **Regionname** using value_counts()` to identify how many expensive houses are located in each region.


```python
expensive_houses = df[( df["Type"] == "h") & (df["Price"]>1000000)]

print("houses (`Type == 'h'`) priced above $1,000,000 : " , len(expensive_houses))

print(expensive_houses["Regionname"].value_counts())
```

## Step 4: 
Use groupby to compute an aggregate statistic per category and interpret the result in a
Markdown cell.

Explanation:
In this step, the dataset was grouped by Regionname using groupby(). Then, the average property price (Price) was calculated for each region using mean(). Finally, the results were sorted in descending order using sort_values(ascending=False) to identify the regions with the highest average property prices.

```python
print(df.groupby("Regionname")["Price"].mean().sort_values(ascending=False))
```
