### Step 1: Why Pandas?

**The problem:** Imagine a manager sends you a CSV file with employee 
data (name, age, salary, department) and asks: what's the average 
salary? How many employees per department? Are there missing values? 
Doing this with plain Python lists means writing loops for every single 
question  slow and repetitive.

**Why isn't NumPy enough on its own?** NumPy makes numeric operations 
very fast, but it only understands raw numbers arranged in a matrix. It 
has no concept of column names  it can't tell that one column is "Age" 
and another is "Salary". That makes it a poor fit for real-world tabular 
data with mixed types (text and numbers together).

**Where Pandas fits in:** Pandas is built on top of NumPy, but adds what 
NumPy is missing: labeled columns, row indexes, and the ability to mix 
data types in one table. It's designed specifically for tabular data - 
the format that CSV files, Excel sheets, SQL query results, JSON, and 
APIs typically come in.

**Where the name comes from:** Despite the panda-bear logo, the name 
actually comes from "**panel data**"  an econometrics term for datasets 
with multiple observations over time or across groups. Over time, Pandas 
became the standard tool for tabular data in general, not just panel 
data specifically.

**The two core objects:**
- **Series** : a single labeled column
- **DataFrame** : a full table made of multiple Series (columns) together

**Where Pandas sits:**

Python → NumPy → Pandas → Scikit-learn / TensorFlow / PyTorch

**A typical data project follows this flow:**

Load (CSV/Excel) → Explore → Clean → Feature Engineering → Visualize → Model

Pandas always comes *before* modeling, because a model can't work with 
messy or unstructured data - that cleanup step is exactly what Pandas is 
built for.

### Step 4:  Attribute vs Method

In Pandas objects, we interact with data using attributes and methods.

##### Attribute:
- Stores information about the object
- Does not require parentheses ()

Example:
series.shape


##### Method:
- Performs an operation
- Requires parentheses ()

Example:
series.mean()

```python
s = pd.Series([10, 20, 30], name="Numbers")

# Attribute - no parentheses, just returns existing info
print(s.dtype)

# Method - has parentheses, calculates something
print(s.mean())
```

### The Core Series Attributes

Ten attributes describe a Series without changing or calculating anything new.  
They only report information about the object:

| Attribute | Description |
|-----------|-------------|
| `values` | The data only (usually stored as a NumPy array) |
| `index` | The row labels |
| `dtype` | The data type of the values |
| `name` | The name assigned to the Series |
| `shape` | The dimensions of the Series as a tuple |
| `size` | The total number of elements |
| `ndim` | The number of dimensions (always `1` for a Series) |
| `axes` | The list of axes (only the index for a Series) |
| `empty` | `True` or `False` - whether the Series is empty |
| `hasnans` | `True` or `False` - whether the Series contains missing values |

```python
students = pd.Series(
    [95, 88, 100],
    index=["Ali", "Sara", "Omar"],
    name="Score"
)

print("Series:")
print(students)

# values: the raw data only, no labels - usually returned as a NumPy array
print("\nValues:", students.values)
print("Type of values:", type(students.values))

# index: the row labels (a special pandas Index object, not a plain list)
print("\nIndex:", students.index)
print("Type of index:", type(students.index))

# dtype: the data type stored inside
print("\nData Type:", students.dtype)

# name: the Series' name (useful when it comes from a DataFrame column)
print("\nName:", students.name)

# shape: dimensions as a tuple - (3,) means "3 elements, 1 dimension"
print("\nShape:", students.shape)

# size: total number of elements (just a number, not a tuple)
print("\nSize:", students.size)

# ndim: number of dimensions - always 1 for a Series
print("\nNdim:", students.ndim)

# axes: returns a list containing the index (Series only has one axis: rows)
print("\nAxes:", students.axes)

# empty: True only if the Series has zero elements
print("\nEmpty:", students.empty)

# hasnans: True if there's at least one missing value (NaN) inside
print("\nHasnans:", students.hasnans)
```

#### Seeing `empty` and `hasnans` in Action

These two attributes are easier to understand by comparing an empty 
Series and a normal one, and a Series with missing data vs. one without.

```python
import numpy as np

# An empty Series
empty_series = pd.Series(dtype=int)
print("Empty series:", empty_series.empty)

# A Series with data
full_series = pd.Series([1, 2, 3])
print("Full series:", full_series.empty)

# A Series containing a missing value (NaN)
with_nan = pd.Series([10, 20, np.nan, 30])
print("Has NaN:", with_nan.hasnans)

# A Series with no missing values
without_nan = pd.Series([10, 20, 30])
print("Has NaN:", without_nan.hasnans)
```

### Step 5: Label based Indexing

Pandas Series can be accessed using the labels of its index.

In label based indexing, we select data by the index name,
not by its numerical position.

The index labels can be strings, numbers, dates, or any other hashable values.

```python
import pandas as pd

scores = pd.Series(
    [85, 90, 78, 92],
    index=["Ali", "Sara", "Omar", "Lina"]
)

print(scores)

print(scores["Sara"]) #Access using label
```

### Step 6: Position based Indexing

Position based indexing accesses elements using their integer location.

Positions start from zero, similar to Python lists.

The first element has position 0.

```python
scores = pd.Series(
    [85, 90, 78, 92],
    index=["Ali", "Sara", "Omar", "Lina"]
)
print (scores)
print(" scores[Ali] Means : Give me the value where index label is Ali")
print(scores["Ali"]) 
print("scores.iloc[0] Means : Give me the first element by position ")
print(scores.iloc[0])
```

 ##### .loc[] : Label based Selection

.loc is explicitly used for selecting data using index labels.

It is recommended because it makes the intention clear.

```python
print (scores.loc["Sara"])

print (scores.loc[["Ali","Omar"]])

print (scores.loc["Ali":"Omar"])
```

#### .iloc[] : Position-based Selection

.iloc selects data using integer positions.

It works like normal Python indexing.

```python
print (scores.iloc[1])

print(scores.iloc[[0,2]])

print (scores.iloc[0:2])
```

### Step 7: Boolean Indexing and Filtering Series

Boolean indexing allows us to filter data based on conditions.

A condition creates a Boolean mask containing `True` and `False` values.
Pandas uses this mask to return only the elements that satisfy the condition.

This concept is very important because the same idea is used later when filtering rows in DataFrames during data cleaning and ML preprocessing.

```python
import pandas as pd

scores = pd.Series(
    [85, 60, 90, 45, 70],
    index=["Ali", "Sara", "Omar", "Lina", "Ahmad"]
)

scores
```

#### Creating Boolean Masks

A Boolean mask is a Series with the same index as the original Series.
Each position contains:

- `True` → condition is satisfied
- `False` → condition is not satisfied

The mask itself does not filter data until we use it for indexing.

```python
mask = scores > 70

mask
```

#### Filtering Using Boolean Indexing

We can use the Boolean mask inside square brackets `[]`
to select only the values where the condition is True.

```python
scores[mask]
```

#### Filtering Directly With Conditions

Instead of creating a separate mask variable,
we can write the condition directly inside the brackets.

```python
scores[scores > 70]
```

#### Comparison Operators

Boolean filtering supports different comparison operators:

- `>` Greater than
- `<` Less than
- `>=` Greater than or equal
- `<=` Less than or equal
- `==` Equal
- `!=` Not equal

```python
print("Scores greater than 80:")
print(scores[scores > 80])


print("\nScores less than 60:")
print(scores[scores < 60])


print("\nScores equal to 70:")
print(scores[scores == 70])
```

#### Combining Multiple Conditions

We can combine conditions using:

| Operator | Meaning |
|----------|---------|
| `&` | AND |
| `|` | OR |
| `~` | NOT |

Important:
Each condition must be wrapped with parentheses.

```python
scores[(scores >= 60) & (scores <= 90)]
```

#### Filtering Categorical Data

Boolean indexing is not limited to numeric values.
We can also filter text categories.

```python
departments = pd.Series(
    ["AI", "Web", "AI", "Data", "Web"],
    index=["Ali", "Sara", "Omar", "Lina", "Ahmad"]
)

departments
```

```python
departments[departments == "AI"]
```

#### Using isin() For Multiple Values

The `isin()` method checks whether each value exists inside a list.

It is useful when filtering multiple categories.

```python
departments.isin(["AI", "Data"])
```

```python
departments[departments.isin(["AI", "Data"])]
```

### Step 8: Boolean Indexing and Fancy Indexing in Pandas Series

#### Boolean Indexing

Boolean indexing allows us to filter data based on conditions.

Instead of selecting values by their labels or positions,
we create a boolean mask that contains True/False values.

Only elements where the condition is True will be returned.

Example:
- Select students with scores greater than 80
- Select employees with salaries above a threshold


#### Fancy Indexing

Fancy indexing means selecting multiple elements by providing
a list of labels or positions.

It allows us to extract multiple values at once.

Example:
- Select specific employees
- Select specific indexes from a Series

```python
import pandas as pd

# Create a Series
scores = pd.Series(
    [85, 90, 78, 92, 65],
    index=["Ali", "Sara", "Omar", "Lina", "Ahmad"]
)

print(scores)


# -------------------------------
# Boolean Indexing
# -------------------------------

# Create a boolean mask
high_scores = scores > 80

print(high_scores)


# Use mask to filter values
print(scores[high_scores])


# Direct filtering
print(scores[scores >= 90])


# Multiple conditions
# Students with score between 80 and 95

print(scores[(scores >= 80) & (scores <= 95)])



# -------------------------------
# Fancy Indexing
# -------------------------------

# Select multiple labels
selected_students = scores[
    ["Ali", "Lina", "Ahmad"]
]

print(selected_students)
```

### Step 9: Series Modification and Alignment

#### Series Modification

Pandas Series are mutable.
We can modify existing values,
add new values,
or delete elements.

#### Alignment

One of the strongest features of Pandas is automatic alignment.

When performing operations between Series,
Pandas matches values based on index labels,
not their positions.

This prevents many data analysis mistakes.

```python
import pandas as pd


sales_2025 = pd.Series(
    [100, 200, 300],
    index=["Jan", "Feb", "Mar"]
)


# Modify existing value
sales_2025["Jan"] = 150


# Add new value
sales_2025["Apr"] = 400


print(sales_2025)



# -------------------------------
# Alignment
# -------------------------------


sales_company_A = pd.Series(
    [100,200,300],
    index=["Jan","Feb","Mar"]
)


sales_company_B = pd.Series(
    [50,70,100],
    index=["Feb","Mar","Apr"]
)



# Pandas aligns using labels

total_sales = sales_company_A + sales_company_B


print(total_sales)
```

Pandas supports arithmetic operations directly on Series.

Operations are performed element-wise.

- Addition

- Division


#### Universal Functions (ufuncs)

ufuncs are NumPy functions that operate on every element

Examples:
- np.sqrt()

### Step 11: Missing Values, Reindexing, Apply and astype

Real-world datasets often contain missing values.

Pandas provides tools to detect and handle missing data.

#### Missing Values

Functions:
- isnull()
- notnull()
- fillna()
- dropna()


#### Reindexing

Changes the order of indexes
or adds new indexes.


#### Apply

Apply a custom function to every element.


#### astype

Convert data types.

```python
import pandas as pd


ages = pd.Series(
    [20,25,None,30],
    index=["Ali","Sara","Omar","Lina"]
)


# Detect missing values

print(ages.isnull())


# Fill missing values

ages = ages.fillna(ages.mean())

print(ages)



# Reindex

new_order = ["Lina","Ali","Sara","Omar"]

print(ages.reindex(new_order))



# Apply function

print(
    ages.apply(lambda x: x+1)
)



# Convert datatype

print(
    ages.astype(int)
)
```

### Step 13: What is a DataFrame?

A DataFrame is the main data structure in Pandas used for working with
tabular data.

A DataFrame represents data in a table format similar to:

- Excel spreadsheets
- SQL tables
- CSV files

It consists of:

- Rows → observations / records
- Columns → features / variables
- Index → row labels

Unlike a Series which represents a single labeled column,
a DataFrame represents a complete dataset containing multiple columns.

Example:

Employee Dataset:

| Index | Name | Age | Salary |
|------|------|-----|--------|
| 0 | Ali | 25 | 5000 |
| 1 | Sara | 30 | 7000 |
| 2 | Omar | 28 | 6000 |


In Machine Learning:

- Each row usually represents one sample.
- Each column usually represents one feature.
- The target column represents what we want to predict.

### Series vs DataFrame

#### Series

A Series is a one-dimensional labeled array.

It contains:

- Values
- Index labels


Example:

Age Series:

| Index | Age |
|------|-----|
| Ali | 25 |
| Sara | 30 |


#### DataFrame

A DataFrame is a two-dimensional labeled data structure.

It contains:

- Multiple Series
- Rows
- Columns


Example:

Employee DataFrame:

| Name | Age | Salary |
|-----|-----|--------|
| Ali | 25 | 5000 |
| Sara | 30 | 7000 |


Relationship:

A DataFrame is basically a collection of Series objects

DataFrame = Multiple Series + Common Index

### Step 20: Pandas DataFrame Inspection Methods

After creating or loading a DataFrame, the first step in any Data Analysis
or Machine Learning project is to inspect and understand the dataset.

Data inspection helps us understand:

- Dataset structure
- Number of rows and columns
- Data types
- Missing values
- Statistical information
- Memory usage


#### Attributes vs Methods

| Type | Description | Example |
|------|-------------|---------|
| Attribute | Gives information about the DataFrame without performing an action. No parentheses needed. | `df.shape` |
| Method | Performs an operation and returns a result. Requires parentheses. | `df.head()` |



#### Important DataFrame Inspection Methods


| Method | Description | What It Returns | Why We Use It |
|--------|-------------|-----------------|---------------|
| `df.head()` | Displays the first rows of the DataFrame. | First 5 rows by default. | Quickly preview the dataset and understand its structure. |

| `df.describe(include="object")` | Generates statistical summary for categorical columns. | Count, unique values, top value, frequency. | Analyze text/category columns. |

### Step 21: Pandas DataFrame Selection

Selecting data is one of the most important operations when working with
Pandas DataFrames.

In real-world Data Analysis and Machine Learning projects, we usually need
to select:

- Specific columns (features)
- Specific rows (samples)
- Specific values
- Subsets of data


DataFrame selection allows us to prepare the data before:

- Data cleaning
- Feature engineering
- Model training


---

#### DataFrame Selection Methods


| Selection Type | Syntax | Returns | Purpose |
|---|---|---|---|
| Select one column | `df["column"]` | Series | Select a single feature |
| Select multiple columns | `df[["col1","col2"]]` | DataFrame | Select multiple features |
| Select rows by label | `df.loc[]` | Series/DataFrame | Select using index labels |
| Select rows by position | `df.iloc[]` | Series/DataFrame | Select using numerical positions |
| Select specific value | `df.loc[row,column]` | Single value | Access one cell |
| Slice rows | `df[start:end]` | DataFrame | Select range of rows |


```python
import pandas as pd

# ==========================================================
# Create Example DataFrame
# ==========================================================

employees = pd.DataFrame(
    {
        "Name": ["Ali", "Sara", "Omar", "Lina", "Ahmad"],

        "Age": [25, 30, 28, 22, 35],

        "Department": ["AI", "CS", "SE", "AI", "CS"],

        "Salary": [5000, 7000, 6000, 4500, 8000]
    }
)


print("Original DataFrame:")
print(employees)


# ==========================================================
# 1. Selecting Single Column
# ==========================================================

# Returns one column as Series

print("\nSelecting Age column:")
print(employees["Age"])


# ==========================================================
# 2. Selecting Multiple Columns
# ==========================================================

# Returns DataFrame containing selected columns

print("\nSelecting Name and Salary columns:")
print(employees[["Name", "Salary"]])



# ==========================================================
# 3. Selecting Rows Using loc
# ==========================================================

# loc uses index labels

print("\nFirst row using loc:")
print(employees.loc[0])


# Selecting specific row and column

print("\nSalary of first employee using loc:")
print(employees.loc[0, "Salary"])


# Selecting multiple rows and columns

print("\nRows 0-2 with Name and Salary:")
print(employees.loc[0:2, ["Name", "Salary"]])

# ==========================================================
# 4. Selecting Rows Using iloc
# ==========================================================

# iloc uses numerical positions

print("\nFirst row using iloc:")
print(employees.iloc[0])


# Selecting specific value

# Row position 0
# Column position 3 (Salary)

print("\nSalary using iloc:")
print(employees.iloc[0,3])


# Selecting rows and columns by positions

print("\nFirst 3 rows and first 2 columns:")
print(employees.iloc[0:3,0:2])


# ==========================================================
# 5. DataFrame Slicing
# ==========================================================

# Basic row slicing

print("\nRows from 0 to 2:")
print(employees[0:3])


# loc slicing includes ending index

print("\nloc slicing:")
print(employees.loc[0:3])

# iloc slicing excludes ending position

print("\niloc slicing:")
print(employees.iloc[0:3])
```

### Step 22: Missing Values & Data Cleaning

Real-world data is rarely perfect - it often has missing values, exact 
duplicate rows, and columns stored as the wrong data type. Cleaning is 
usually the largest part of any data project, and everything downstream 
depends on doing it correctly.

**Detecting missing values:** `.isnull()` (same as `.isna()`) returns a 
True/False grid marking every missing cell. Combined with `.sum()`, it 
gives a quick count of missing values per column.

**Handling missing values - two strategies:**
- `.dropna()` removes any row containing at least one missing value. 
  Simple, but you lose data - only safe when missing rows are rare.
- `.fillna(value)` replaces missing values instead of removing rows. The 
  replacement value depends on the column type:
  - Numeric columns: fill with the **mean** or **median**
  - Categorical columns: fill with the **mode** (most frequent value) or 
    a constant like `"Unknown"`
  - Time-ordered data: `.ffill()` (carries the last known value forward) 
    or `.bfill()` (pulls the next known value backward)

**Duplicates:** `.duplicated()` flags rows that are exact copies of an 
earlier row (True = duplicate). `.drop_duplicates()` removes them, 
keeping only the first occurrence by default.

**Type conversion:** `.astype(type)` converts a column to a different 
data type (e.g., int to float) - important because some operations 
require a specific type to work correctly.

**Replacing specific values:** `.replace(old, new)` swaps out a specific 
value everywhere it appears in a column - useful for standardizing 
inconsistent category labels (e.g., "AI" → "Artificial Intelligence").

```python
import pandas as pd
import numpy as np

# A small dataset with an intentional missing Age, missing Department,
# and one exact duplicate row (Ali appears twice, identically)
employees = pd.DataFrame({
    "Name": ["Ali", "Sara", "Omar", "Lina", "Ahmad", "Noor", "Ali"],
    "Age": [25, 30, 28, 22, 35, np.nan, 25],
    "Department": ["AI", "CS", "SE", "AI", "CS", None, "AI"],
    "Salary": [5000, 7000, 6000, 4500, 8000, 5500, 5000]
})
print("Original data:")
print(employees)

# --- 1. Detecting missing values ---
# isnull() returns True/False for every cell; .sum() counts True per column
print("\nMissing values per column:")
print(employees.isnull().sum())

# --- 2. dropna() - remove any row with at least one missing value ---
print("\nAfter dropna():")
print(employees.dropna())

# --- 3. fillna() - fill instead of removing ---
# Numeric column -> fill with the mean
age_filled = employees["Age"].fillna(employees["Age"].mean())
print("\nAge filled with mean:")
print(age_filled)

# Categorical column -> fill with the mode (most common value)
dept_filled = employees["Department"].fillna(employees["Department"].mode()[0])
print("\nDepartment filled with mode:")
print(dept_filled)

# Or fill with a constant placeholder instead
print("\nDepartment filled with a constant:")
print(employees["Department"].fillna("Unknown"))

# Forward fill / backward fill - useful for ordered/time-series data
print("\nAge with forward fill:")
print(employees["Age"].ffill())
print("\nAge with backward fill:")
print(employees["Age"].bfill())

# --- 4. Duplicates ---
print("\nWhich rows are duplicates:")
print(employees.duplicated())

print("\nAfter drop_duplicates():")
print(employees.drop_duplicates())

# --- 5. Type conversion ---
print("\nSalary converted to float:")
print(employees["Salary"].astype(float))

# --- 6. Replacing specific values ---
print("\nStandardizing 'AI' to a full label:")
print(employees["Department"].replace("AI", "Artificial Intelligence"))
```

### Step 23: GroupBy - Split, Apply, Combine

`groupby()` is one of the most powerful Pandas operations. The idea is 
"split-apply-combine": split the data into groups based on a column's 
values, apply an aggregation (mean, sum, count...) to each group 
separately, then combine the results into one summary.

**Basic groupby:** `df.groupby("column")["other_column"].mean()` groups 
rows by a category and computes the mean for each group.

**`.agg()`:** Instead of one statistic, `.agg([...])` computes several 
at once (mean, min, max, count...) in a single call. You can also pass a 
dictionary to apply a *different* aggregation to each column.

**Grouping by multiple columns:** Passing a list `["col1", "col2"]` 
groups by the combination of both columns - useful for more detailed 
breakdowns (e.g., department AND seniority level).

**`.size()` vs. `.count()`:** `.size()` counts all rows in each group 
(including missing values). `.count()` counts only non-missing values in 
a specific column - the two can differ if that column has NaNs.

**`.transform()`:** Unlike `.groupby().mean()` (which shrinks the data 
down to one row per group), `.transform()` broadcasts the group result 
back to match the original DataFrame's shape - useful for adding a "group 
average" column right next to the original data.

**`.filter()`:** Keeps or drops entire groups based on a condition 
applied to the group as a whole (e.g., "keep only departments whose 
average salary is above some threshold").

```python
import pandas as pd

employees = pd.DataFrame({
    "Name": ["Ali", "Sara", "Omar", "Lina", "Ahmad", "Noor"],
    "Age": [25, 30, 28, 22, 35, 27],
    "Department": ["AI", "CS", "SE", "AI", "CS", "SE"],
    "Salary": [5000, 7000, 6000, 4500, 8000, 5500]
})
print(employees)

# --- 1. Basic groupby: average salary per department ---
print("\nAverage salary per department:")
print(employees.groupby("Department")["Salary"].mean())

# --- 2. agg(): multiple statistics at once ---
print("\nMultiple stats per department:")
print(employees.groupby("Department")["Salary"].agg(["mean", "min", "max", "count"]))

# --- 3. Different aggregation for different columns ---
print("\nMean salary AND max age per department:")
print(employees.groupby("Department").agg({"Salary": "mean", "Age": "max"}))

# --- 4. Grouping by multiple columns ---
employees["Level"] = ["Junior", "Senior", "Junior", "Junior", "Senior", "Senior"]
print("\nAverage salary per department AND level:")
print(employees.groupby(["Department", "Level"])["Salary"].mean())

# --- 5. size() vs count() ---
print("\nsize() - counts all rows per group:")
print(employees.groupby("Department").size())
print("\ncount() - counts non-missing Salary values per group:")
print(employees.groupby("Department")["Salary"].count())

# --- 6. transform(): add the group average back onto every row ---
employees["dept_avg_salary"] = employees.groupby("Department")["Salary"].transform("mean")
print("\nWith department average added as a new column:")
print(employees)

# --- 7. filter(): keep only groups matching a condition ---
print("\nOnly departments with average salary above 5000:")
print(employees.groupby("Department").filter(lambda x: x["Salary"].mean() > 5000))
```
