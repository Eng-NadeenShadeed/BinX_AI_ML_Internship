##  Why Visualization Matters

Data in tabular form can hide important patterns and relationships between variables. Visualizations make it much easier to explore and understand data by helping us:

- Identify trends over time.
- Discover relationships between variables.
- Detect outliers and unusual observations.
- Communicate insights more effectively.

For these reasons, data visualization is an essential part of **Exploratory Data Analysis (EDA)** and presenting data analysis results.

#### Figure

A **Figure** is the entire plotting window or canvas.

A single figure can contain one or multiple plots.

````python
fig = plt.figure()
plt.show()
````

#### Axes

An **Axes** represents an individual plot inside the figure.

When using `plt.subplots()`, each subplot is an Axes object.

````python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.plot([1, 2, 3], [2, 4, 6])

plt.show()
````

#### Axis

````python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3], [4, 5, 6])

plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
````

### Step 5: Types of Plots in Matplotlib

Matplotlib provides different types of plots for different data visualization tasks. Choosing the appropriate plot depends on the type of data and the insights you want to communicate.

#### 1. Line Plot

A **Line Plot** is used to show changes or trends over a continuous variable, such as time.

#### Common Uses

- Stock prices
- Temperature changes
- Monthly sales
- Population growth

#### Syntax

```python
plt.plot(x, y)
```

````python
months = ["Jan", "Feb", "Mar", "Apr"]
sales = [100, 120, 180, 250]

plt.plot(months, sales)

plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.show()
````

#### 2. Scatter Plot

A **Scatter Plot** is used to visualize the relationship between two numerical variables.

#### Common Uses

- Study hours vs. exam scores
- Height vs. weight
- Age vs. salary

#### Syntax

```python
plt.scatter(x, y)
```

````python
hours = [1, 2, 3, 4, 5]
marks = [55, 60, 70, 80, 95]

plt.scatter(hours, marks)

plt.title("Study Hours vs Marks")
plt.xlabel("Hours")
plt.ylabel("Marks")

plt.show()
````

#### 3. Bar Chart

A **Bar Chart** is used to compare values across different categories.

#### Common Uses

- Sales by product
- Number of students in each department
- Number of cars by brand

#### Syntax

```python
plt.bar(x, y)
```

````python
import matplotlib.pyplot as plt

fruits = ["Apple", "Banana", "Orange"]
sales = [30, 50, 20]

plt.bar(fruits, sales)

plt.title("Fruit Sales")
plt.xlabel("Fruit")
plt.ylabel("Sales")

plt.show()
````

#### 4. Histogram

A **Histogram** is used to display the distribution of numerical data by grouping values into intervals called **bins**.

#### Common Uses

- Age distribution
- Exam score distribution
- Salary distribution
- Detecting outliers

#### Syntax

```python
plt.hist(x, bins=10)
```

> **Note:** `bins` specifies the number of intervals used to group the data.

````python
ages = [18, 19, 20, 20, 21, 21, 21, 22, 22, 23, 24, 25, 26]

plt.hist(ages, bins=5)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
````

#### 5. Pie Chart

A **Pie Chart** is used to show the proportion or percentage of each category as part of a whole.

#### Common Uses

- Budget distribution
- Market share
- Programming language popularity

#### Syntax

```python
plt.pie(values, labels=labels, autopct="%1.1f%%")
```

````python
languages = ["Python", "Java", "C++"]
students = [50, 30, 20]

plt.pie(students, labels=languages, autopct="%1.1f%%")

plt.title("Programming Languages")

plt.show()
````

#### 6. Box Plot

A **Box Plot** summarizes the distribution of numerical data and helps identify outliers.

It displays:

- Minimum
- First Quartile (Q1)
- Median
- Third Quartile (Q3)
- Maximum

#### Syntax

```python
plt.boxplot(data)
```

````python
scores = [55, 60, 62, 65, 70, 72, 75, 90, 95]

plt.boxplot(scores)

plt.title("Scores Distribution")
plt.ylabel("Score")

plt.show()
````

#### 7. Heatmap

A **Heatmap** represents values in a matrix using colors.

#### Common Uses

- Correlation matrices
- Image visualization
- Density maps

#### Syntax

```python
plt.imshow(data, cmap="viridis")

plt.colorbar()
```

````python
import numpy as np

data = np.random.rand(5, 5)

plt.imshow(data, cmap="viridis")

plt.colorbar()

plt.title("Heatmap")

plt.show()
````

### Step 6: Subplots

A **subplot** allows you to display multiple plots within the same figure.

Instead of creating separate windows for each chart, subplots organize multiple visualizations into a single figure, making it easier to compare different datasets.

Subplots are commonly used in data analysis dashboards and exploratory data analysis (EDA).

---

#### Creating Subplots

The `plt.subplots()` function is used to create a figure and one or more axes.

#### Syntax

```python
fig, ax = plt.subplots(rows, columns)
```

#### Parameters

- **rows**: Number of subplot rows.
- **columns**: Number of subplot columns.

The function returns:

- **fig** → the entire figure (canvas).
- **ax** → the axes object(s) where plots are drawn.

---


##### Single Subplot Example

````python
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

fig, ax = plt.subplots()

ax.plot(x, y)

ax.set_title("Line Plot")
ax.set_xlabel("X")
ax.set_ylabel("Y")

plt.show()
````

##### Multiple Subplots Example

The following example creates two plots in one figure.

````python
x = [1, 2, 3, 4]

y1 = [2, 4, 6, 8]
y2 = [8, 6, 4, 2]

fig, ax = plt.subplots(1, 2, figsize=(10, 4))

ax[0].plot(x, y1)
ax[0].set_title("Increasing")

ax[1].plot(x, y2)
ax[1].set_title("Decreasing")

plt.show()
````

##### 2 × 2 Subplots Example

````python
fig, ax = plt.subplots(2, 2, figsize=(8, 6))

ax[0, 0].plot([1, 2, 3], [2, 4, 6])
ax[0, 0].set_title("Plot 1")

ax[0, 1].scatter([1, 2, 3], [3, 5, 4])
ax[0, 1].set_title("Plot 2")

ax[1, 0].bar(["A", "B", "C"], [4, 7, 5])
ax[1, 0].set_title("Plot 3")

ax[1, 1].hist([1, 2, 2, 3, 3, 3, 4], bins=4)
ax[1, 1].set_title("Plot 4")

plt.tight_layout()

plt.show()
````

#### `plt.tight_layout()`

When multiple subplots are displayed, titles and labels may overlap.

The `plt.tight_layout()` function automatically adjusts the spacing between subplots to improve readability.

##### Syntax

```python
plt.tight_layout()
```
