## Object-Oriented Programming (OOP)

In this section, I document what I learned about OOP: how classes work 
as blueprints, how objects are independent instances created from them, 
and the building blocks (`__init__`, `self`, attributes, and methods) 
that make this possible.

### 1. Class vs. Object

A **class** is a blueprint or template - it defines what attributes and 
methods objects of this type will have, but it doesn't hold any real data 
by itself. An **object** is an actual instance created from that class.

Analogy: a class is like a cookie cutter, and an object is an actual 
cookie made using it. The shape (structure) is the same, but each cookie 
is a separate, real thing.

```python
class Intern: # intern class to represent an intern in the program
    pass  #pass is a placeholder that indicates that the class has no attributes or methods defined yet

intern1 = Intern() # intern1 is an instance of the Intern class(object)
print(intern1)
print(type(intern1))
```

### 2. `__init__` and `self`

`__init__` is a special method that runs automatically whenever a new 
object is created from the class. It's used to set up the object's 
initial data.

`self` refers to "this specific object" - the one currently being created 
or used. Writing `self.name = name` means "store this value inside this 
particular object", not inside the class itself or any other object.

```python
class Intern:
    def __init__(self , name , hours_completed):
        self.name = name
        self.hours_completed = hours_completed

intern1 = Intern("Nadeen" , 80)
print (intern1.name)
print (intern1.hours_completed)
```

### 3. Methods

While attributes store an object's data, methods define what an object 
can *do*. A method is just a function defined inside a class, and it 
always takes `self` as its first parameter so it can access the object's 
own data.

```python
class Intern:
    def __init__(self, name, hours_completed, total_hours=400):
        self.name = name
        self.hours_completed = hours_completed
        self.total_hours = total_hours

    def progress_report(self):
        percent = (self.hours_completed / self.total_hours) * 100
        print(f"{self.name}: {self.hours_completed}/{self.total_hours} hours ({percent:.1f}%)")

    def is_eligible_for_certificate(self, final_grade, attendance_percentage):
        return final_grade >= 70 and attendance_percentage >= 90

intern1 = Intern("Nadeen", 80)
intern1.progress_report()
print(intern1.is_eligible_for_certificate(78, 92))
print(intern1.is_eligible_for_certificate(65, 92))
```

### 4. Objects Are Independent

Each object created from a class has its own separate copy of the 
instance attributes. Changing one object's data has no effect on any 
other object created from the same class - even though they share the 
same blueprint.

```python
class Intern:
    def __init__(self, name, hours_completed, total_hours=400):
        self.name = name
        self.hours_completed = hours_completed
        self.total_hours = total_hours

    def progress_report(self):
        percent = (self.hours_completed / self.total_hours) * 100
        print(f"{self.name}: {self.hours_completed}/{self.total_hours} hours ({percent:.1f}%)")

    def is_eligible_for_certificate(self, final_grade, attendance_percentage):
        return final_grade >= 70 and attendance_percentage >= 90
    
intern1 = Intern("Nadeen", 80)
intern2 = Intern("Jenan" , 90)
print("before changing intern1's hours_completed:")
intern1.progress_report()
intern2.progress_report()

# Changing intern1 does not affect intern2 at all
print("\nAfter changing intern1's hours_completed:")
intern1.hours_completed = 100
intern1.progress_report()
intern2.progress_report()  # intern2's hours_completed remains unchanged
```

### 5. Class Variables vs. Instance Variables

A **class variable** is defined directly inside the class (outside 
`__init__`) and holds the same value for every object created from that 
class. An **instance variable** is defined using `self` inside `__init__` 
and holds a unique value for each individual object.

```python
class Intern:
    program_name = "BinX AI & ML Internship"   # class variable - shared

    def __init__(self, name, hours_completed):
        self.name = name                        # instance variable - unique
        self.hours_completed = hours_completed

intern1 = Intern("Nadeen", 16)
intern2 = Intern("Jenan", 120)

print(intern1.program_name)
print(intern2.program_name)
print(intern1.name, intern2.name)
```

### 1. Inheritance

Inheritance lets a new class (the "child") automatically reuse the 
attributes and methods of an existing class (the "parent"), while also 
adding its own extra behavior. This avoids rewriting the same code in 
multiple classes.

I modeled this using a base `MLModel` class, with `ClassificationModel` 
and `RegressionModel` inheriting from it - similar to how different model 
types will work later in this program (Phase 2: Supervised Learning).

```python
class MLModel:
    def __init__(self, name):
        self.name = name

    def describe(self):
        print(f"This is a {self.name} model")

class ClassificationModel(MLModel):
    def predict(self):
        print("Predicting a category (e.g., churn: yes/no)")

class RegressionModel(MLModel):
    def predict(self):
        print("Predicting a number (e.g., house price)")

clf = ClassificationModel("Classification")
clf.describe()   # inherited from MLModel
clf.predict()    # defined in ClassificationModel itself

reg = RegressionModel("Regression")
reg.describe()
reg.predict()
```

### 2. Polymorphism

Polymorphism means the same method name can behave differently depending 
on which object calls it. I looped over a list containing both a 
`ClassificationModel` and a `RegressionModel`, calling `.predict()` on 
each - the same method name produced different output for each object, 
without needing to check the object's type manually.

```python
models = [ClassificationModel("Classification"), RegressionModel("Regression")]
for model in models:
    model.predict()
```

### 3. Encapsulation

Encapsulation means hiding an object's internal data so it can't be 
accessed or changed directly from outside the class - access must go 
through a defined method instead. Prefixing an attribute with `__` marks 
it as private in Python.

I used this to protect a `random_seed` value, since accidentally changing 
it later in a project would break reproducibility (a core rule mentioned 
throughout this program).

```python
class TrainingConfig:
    def __init__(self, seed):
        self.__random_seed = seed   # private attribute

    def get_seed(self):
        return self.__random_seed

config = TrainingConfig(42)
print(config.get_seed())        # correct way to access it

#print(config.__random_seed)     # direct access attempt - this will fail
```

### 4. Abstraction

Abstraction defines a required structure ("contract") that every subclass 
must follow, without specifying exactly how each one implements it. Using 
`ABC` and `@abstractmethod`, I created a `Model` blueprint that forces any 
subclass to implement a `train()` method - but each subclass is free to 
implement it differently.

This mirrors how Scikit-learn's `BaseEstimator` requires every model class 
to implement `.fit()` and `.predict()`, even though each model type 
(linear regression, decision trees, etc.) trains completely differently 
under the hood.

```python
from abc import ABC, abstractmethod # this import is necessary to use abstract base classes and abstract methods

class Model(ABC):
    @abstractmethod # this decorator indicates that the method is abstract and must be implemented by subclasses
    def train(self):
        pass

class LinearModel(Model):
    def train(self):
        print("Training a linear model using gradient descent")

class TreeModel(Model):
    def train(self):
        print("Training a decision tree by splitting nodes")

lm = LinearModel()
lm.train()

tm = TreeModel()
tm.train()
```
