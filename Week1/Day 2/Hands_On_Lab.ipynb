Step 1: Write a function that takes a list of numbers and returns the mean, min, and max as a dictionary.

```python
def grade_stats(grade):
    stats ={
        "mean": sum(grade) / len(grade),
        "min": min(grade),
        "max": max(grade)
    }
    return stats
Weekly_Scores = {78 , 82 , 85 , 90}
result = grade_stats(Weekly_Scores)
print(result)
```

## Step 2
Rewriting a for-loop that filters even numbers as a single list comprehension.

```python
# original using for loop
print("Even numbers from 1 to 20 using for loop:")
even_number_for = [ ]

for i in range(1, 21):
    if i % 2 == 0:
        even_number_for.append(i)
print(even_number_for)
print("Even numbers from 1 to 20 using list comprehension:")
# using list comprehension
even_number = [x for x in range(1, 21) if x % 2 == 0]
print(even_number)
```

## Step 3
A small class representing a data record with two attributes and one method.

```python
class Student:
    def __init__(self, name, age, grade):
        self.name = name
        self.age = age
        self.grade = grade

    def display_info(self):
        print(f"Name: {self.name}, Age: {self.age}, Grade: {self.grade}")

student1  = Student("Nadeen", 21, "A")
student1.display_info()
```
